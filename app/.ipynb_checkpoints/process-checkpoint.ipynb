```python
import pandas as pd 
import re
import nltk
from nltk import ngrams, FreqDist
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('all_subs.csv')
df1 = pd.DataFrame(data)
```

```python
df1.head()
```

```python
all_subs = df1['content'].values.tolist()
```

```python
all_subs[3]
```

```python
main_str = '. '.join(all_subs)
```

```python
main_str
```

```python

# replace all square bracket contents
pattern = r'\[.*?\]'
signs = re.sub(pattern, '',main_str)
```

```python
signs
```

```python
text_token = nltk.tokenize.word_tokenize(signs)
```

```python
stop_list = nltk.corpus.stopwords.words('english')
```

```python
text_token_1 = [x.lower() for x in text_token if x not in stop_list and x != '.']
```

```python
text_token_1
```

```python
all_counts = dict()
for size in range(2,4):
    all_counts[size] = FreqDist(ngrams(text_token_1, size))
```

```python
all_counts[2].most_common(10)
```

```python
names = [' '.join(x[0]) for x in all_counts[2].most_common(20)]
values = [int(x[1]) for x in all_counts[2].most_common(20)]
```

```python
# remove all bigrams
join_text = ' '.join(text_token_1)
join_text = join_text.replace('magna laude', ' ')
join_text = join_text.replace('summa laude', ' ')
```

```python
text_token_2 = [x.lower() for x in nltk.tokenize.word_tokenize(join_text)]
all_counts2 = dict()
for size in range(1,4):
    all_counts2[size] = FreqDist(ngrams(text_token_2, size))
```

```python
all_counts2[1].most_common(10)
```

```python
# add unigram to bigrams
names1 = [x[0][0] for x in all_counts2[1].most_common(20)] + names
values1 = [int(x[1]) for x in all_counts2[1].most_common(20)] + values
```

```python
plt.figure(figsize = [20,7])
plt.bar(names1, values1)
plt.xticks(list(range(len(names1))), names1, rotation = 90)
# plt.axes
plt.show()
```

```python
all_data_cleaned = []
for i in range(len(names1)):
    all_data_cleaned.append((names1[i], values1[i]))
```

```python
# show all data

print(all_data_cleaned)
```
